```python
# install beautiful soup
# install lxml (parser)
# install html5lib (parser)
# install requests

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Links of all pages

links = []
def get_link():
    for i in range(1,238):
        links.append("https://www.consumeraffairs.com/online/amazon.html?page="+str(i))

# Call function        
get_link()
```

```python
links
```

```python
# Function to get data from all pages

# list for rating
rating_list = []


# list for review
review_list = []


def get_data():
    for i in links:
        source = requests.get(i).text
        soup = BeautifulSoup(source,'lxml')
        
        # Extracting reviews of customers
        for article in soup.find_all('div',class_='rvw js-rvw'):
            review = article.find('div', class_='rvw-bd')
            review_list.append(review.text)
            
        # Extracting rating of customers
        for article1 in soup.find_all('div',class_='rvw__hdr-stat'):
            #rating = 
            rating_list.append(article1.find('meta',itemprop='ratingValue')['content'])

# Calling function
get_data()
```

```python
len(rating_list)
```

```python
# Creating Dataframe and CSV file

reviews = pd.DataFrame(review_list)
rating = pd.DataFrame(rating_list)

data = pd.concat([reviews,rating], axis=1)
data.to_csv('Reviews.csv')
```

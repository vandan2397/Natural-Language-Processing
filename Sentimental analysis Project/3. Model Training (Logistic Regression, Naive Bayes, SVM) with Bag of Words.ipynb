<h3>Import Libraries and Data</h3>

```python
# import libraries

import pandas as pd
```

```python
# import pre-processed reviews and labels

# Reviews
reviews  = pd.read_csv('/content/drive/MyDrive/Sentiment analysis/Natural-Language-Processing-main/Sentimental analysis Project/Data/Pre-Processed.csv')
reviews.drop(['Unnamed: 0'],axis=1,inplace=True)

# Labels
labels = pd.read_csv('/content/drive/MyDrive/Sentiment analysis/Natural-Language-Processing-main/Sentimental analysis Project/Data/Amazon_Customer_Reviews.csv')
```

<h3> Label Coding</h3>

```python
# dummyfying output variable

labels = pd.get_dummies(labels['Ratings'],drop_first=True)
```

<h3>Train-Test Split</h3>

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(reviews, labels, test_size = 0.20, random_state = 42)
```

```python
# Appending each review to list, so that it becomes iterable and compatible for count vectorizer

X_train_list = []
X_test_list =[]

for index, row in X_train.iterrows():
  X_train_list.append(row['Reviews'])

for index, row in X_test.iterrows():
  X_test_list.append(row['Reviews'])
```

<h3>Bag of Words Model</h3>

```python
# Creating the Bag of Words model

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000)
```

```python
# Vectorizing train reviews (Bag of Words)

X_train_bow = vectorizer.fit_transform(X_train_list).toarray()
```

```python
# Vectorizing test reviews (Bag of Words)

X_test_bow = vectorizer.fit_transform(X_test_list).toarray()
```

```python
# Shape of Train and Test

print("Train Set: ", X_train_bow.shape)
print("Test Set: ", X_test_bow.shape)
```

```python
# Save Count vectorizer
# from sklearn.externals import joblib

import pickle

pickle.dump(vectorizer, open('tranform.pkl', 'wb'))
```

<h3>Naive Bayes Classifier</h3>

```python
# Training model using Naive bayes classifier

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB(alpha=0.01).fit(X_train_bow, y_train)
y_pred=model.predict(X_test_bow)
```

```python
# Training Evaluation

y_pred_train = model.predict(X_train_bow)

# Classification metrics
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

cm_train = confusion_matrix(y_pred_train, y_train)
print("Confusion Matrix:")
print(cm_train)

print("____________________________________________________________")

report_train = classification_report(y_pred_train, y_train)
print("Classification Report:",)
print (report_train)

print("____________________________________________________________")

accuracy_train = accuracy_score(y_pred_train,y_train)
print("Accuracy:",accuracy_train)
```

```python
# Testing Evaluation Evaluation

y_pred_test = model.predict(X_test_bow)

# Classification metrics
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
cm_test = confusion_matrix(y_pred, y_test)
print("Confusion Matrix:")
print(cm_test)

print("____________________________________________________________")

report_test = classification_report(y_pred, y_test)
print("Classification Report:",)
print (report_test)

print("____________________________________________________________")

accuracy_test = accuracy_score(y_pred,y_test)
print("Accuracy:",accuracy_test)
```
